```python
import numpy as np
import pandas as pd

#!pip install xgboost
import xgboost as xgb

from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score

#!pip install shap
import matplotlib.pyplot as plt
```

```python
Plot_Results=pd.read_csv('output/model2.csv')
```

```python
Plot_Results.sort_values(by='param_n_estimators', inplace=True)

# Find values of AUC for learning rate of 0.05 and different values of depth
lr_d2 = Plot_Results.loc[(Plot_Results['param_learning_rate']==0.05) & (Plot_Results['param_num_parallel_tree']==1),:]
lr_d3 = Plot_Results.loc[(Plot_Results['param_learning_rate']==0.05) & (Plot_Results['param_num_parallel_tree']==3),:]
lr_d5 = Plot_Results.loc[(Plot_Results['param_learning_rate']==0.05) & (Plot_Results['param_num_parallel_tree']==5),:]


# Let us plot now
fig, ax = plt.subplots(figsize=(10,5))
lr_d2.plot(x='param_n_estimators', y='mean_test_score', label='PTree=1', ax=ax)
lr_d3.plot(x='param_n_estimators', y='mean_test_score', label='PTree=3', ax=ax)
lr_d5.plot(x='param_n_estimators', y='mean_test_score', label='PTree=5', ax=ax)

plt.ylabel('Mean Validation AUC')
plt.title('Performance wrt # of Trees and Parallel Trees')
```
